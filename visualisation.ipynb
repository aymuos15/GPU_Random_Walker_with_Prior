```python
import numpy as np
import torch

from helper_numpy import random_walker_prior    
from helper_torch import random_walker_prior_torch

import matplotlib.pyplot as plt

# Create the test image with two squares
a = np.zeros((50, 50))
a[10:20, 10:20] = 1
a[30:40, 30:40] = 1
a += 0.7 * np.random.random((50, 50))

# Create prior probabilities based on image intensity
p = (1 - a) ** 2  # Higher probability for background (class 1)
q = a ** 2        # Higher probability for foreground (class 2)
prior = np.array([p.ravel(), q.ravel()])

# Perform segmentation (random_walker_prior function is assumed to be implemented)
labs = random_walker_prior(a, prior)

# Perform segmentation using PyTorch
a = torch.tensor(a)
prior = torch.tensor(prior)
labs_torch = random_walker_prior_torch(a, prior)

# Visualization
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(15, 5))

# Original image
im1 = ax1.imshow(a, cmap='gray')
ax1.set_title('Original Image')
plt.colorbar(im1, ax=ax1)

# Prior probabilities
prior_vis = prior[1].reshape(a.shape)  # Visualize probability of second class
im2 = ax2.imshow(prior_vis, cmap='viridis')
ax2.set_title('Prior Probability (Class 2)')
plt.colorbar(im2, ax=ax2)

# Segmentation result
im3 = ax3.imshow(labs, cmap='coolwarm')
ax3.set_title('Segmentation Result')
plt.colorbar(im3, ax=ax3)

# Segmentation result using PyTorch
im4 = ax4.imshow(labs_torch.cpu().numpy(), cmap='coolwarm')
ax4.set_title('Segmentation Result (PyTorch)')
plt.colorbar(im4, ax=ax4)

plt.tight_layout()
plt.show()

# Check if the results are the same
print("Are the results same?:", np.allclose(labs, labs_torch.cpu().numpy()))
```
